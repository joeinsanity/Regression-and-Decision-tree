```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import learning_curve
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Fetch the dataset from OpenML using data ID 314
dataset = fetch_openml(data_id=314)

# Convert the dataset to a DataFrame
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
target = dataset.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.2, random_state=42)

# Specify the indices of the nominal features
nominal_indices = [2, 3]

# Define the ColumnTransformer for one-hot encoding
ct = ColumnTransformer([('encoder', OneHotEncoder(), nominal_indices)], remainder='passthrough')

# Apply one-hot encoding to the training and testing sets
X_train_encoded = ct.fit_transform(X_train)
X_test_encoded = ct.transform(X_test)

# Define the k values to compare
k_values = [5, 10, 15]

# Initialize an empty list to store rmse values
rmse_values = []

# Define the size of the training set
train_sizes = np.linspace(0.1, 1.0, 10)

# Iterate over each k value
for k in k_values:
    # Initialize k-nearest neighbor regressor
    knn_regressor = KNeighborsRegressor(n_neighbors=k)

    # Compute learning curve
    train_sizes_abs, train_scores, test_scores = learning_curve(
        knn_regressor, X_train_encoded, y_train, train_sizes=train_sizes, cv=10,
        scoring='neg_root_mean_squared_error')

    # Compute rmse for the last point of the learning curve
    last_rmse = np.sqrt(-test_scores[:, -1])
    rmse_values.append(last_rmse)

# Plot learning curves for each k value
plt.figure()
for i, k in enumerate(k_values):
    plt.plot(train_sizes_abs, rmse_values[i], label=f'k={k}')

plt.title('Learning Curves for k-Nearest Neighbor Regressor')
plt.xlabel('Number of Training Examples')
plt.ylabel('Root Mean Squared Error')
plt.legend()
plt.show()

# Display the rmse values in a table
rmse_table = pd.DataFrame(rmse_values, index=k_values, columns=train_sizes_abs)
print("RMSE values for k-Nearest Neighbor Regressor with different k values:")
print(rmse_table)
```

```python



## basically the same code
X = data.data
y = data.target

y = y.astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

param_grid = {'min_samples_leaf': [1, 5, 10, 15, 20]}

entropy_tree = DecisionTreeClassifier(criterion='entropy')
gini_tree = DecisionTreeClassifier(criterion='gini')

grid_search_entropy = GridSearchCV(entropy_tree, param_grid=param_grid, cv=10, scoring='roc_auc')
grid_search_gini = GridSearchCV(gini_tree, param_grid=param_grid, cv=10, scoring='roc_auc')

grid_search_entropy.fit(X_train, y_train)
grid_search_gini.fit(X_train, y_train)

best_entropy_params = grid_search_entropy.best_params_
best_gini_params = grid_search_gini.best_params_

entropy_tree = DecisionTreeClassifier(criterion='entropy', **best_entropy_params)
gini_tree = DecisionTreeClassifier(criterion='gini', **best_gini_params)

cv_scores_entropy = cross_val_score(entropy_tree, X_train, y_train, cv=10, scoring='roc_auc')
cv_scores_gini = cross_val_score(gini_tree, X_train, y_train, cv=10, scoring='roc_auc')

# Compute ROC curve and AUC for final models
fpr_entropy, tpr_entropy, _ = roc_curve(y_test, entropy_tree.fit(X_train, y_train).predict_proba(X_test)[:, 1])
roc_auc_entropy = auc(fpr_entropy, tpr_entropy)

fpr_gini, tpr_gini, _ = roc_curve(y_test, gini_tree.fit(X_train, y_train).predict_proba(X_test)[:, 1])
roc_auc_gini = auc(fpr_gini, tpr_gini)

# plot entropy
from matplotlib import pyplot as plt
plt.xlabel("1 - Specificity")
plt.ylabel("Sensitivity")
plt.xlim(0,1)
plt.ylim(0,1)
plt.plot(fpr_entropy,tpr_entropy,label="Entropy Decision Tree")
plt.legend()
plt.show()
##plot gini
plt.xlabel("1 - Specificity")
plt.ylabel("Sensitivity")
plt.xlim(0,1)
plt.ylim(0,1)
plt.plot(fpr_gini,tpr_gini,label="gini Decision Tree")
plt.legend()
plt.show()
print("Entropy best Params:", best_entropy_params)
print("Gini best Params:", best_gini_params)
print("Cross validation AUC scores (Entropy):", cv_scores_entropy)
print("Cross validation AUC scores (Gini):", cv_scores_gini)
```
